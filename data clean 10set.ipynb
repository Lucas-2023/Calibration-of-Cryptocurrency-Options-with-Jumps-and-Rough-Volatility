```python
# Cell 1
# Check current folder and files

import os
import zipfile
import glob
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

print("Current working directory:")
print(os.getcwd())

print("\nFiles in this folder:")
for f in os.listdir():
    print(f)
```

```python
# Cell 2
# Define daily zip files and BTC underlying prices

zip_info = {
    "20260422.zip": "2026-04-22",
    "20260423.zip": "2026-04-23",
    "20260426.zip": "2026-04-26",
    "20260427.zip": "2026-04-27",
    "20260428.zip": "2026-04-28",
    "20260429.zip": "2026-04-29",
    "20260501.zip": "2026-05-01",
    "20260502.zip": "2026-05-02",
    "20260503.zip": "2026-05-03",
    "20260504.zip": "2026-05-04",
}

underlying_price_map = {
    "2026-04-22": 78203,
    "2026-04-23": 78268,
    "2026-04-26": 78657,
    "2026-04-27": 77335,
    "2026-04-28": 76408,
    "2026-04-29": 76226,
    "2026-05-01": 78372,
    "2026-05-02": 78170,
    "2026-05-03": 79254,
    "2026-05-04": 80081,
}

print("Number of zip files:", len(zip_info))
print("Zip files:")
for zip_file, date_label in zip_info.items():
    print(zip_file, "->", date_label)
```

```python
# Cell 3
# Unzip each daily zip file into a folder

def unzip_file(zip_path, extract_to):
    os.makedirs(extract_to, exist_ok=True)
    with zipfile.ZipFile(zip_path, "r") as z:
        z.extractall(extract_to)
    print(f"Unzipped {zip_path} -> {extract_to}")

folders = {}

for zip_file, date_label in zip_info.items():
    folder_name = "day_" + date_label.replace("-", "")
    folders[date_label] = folder_name
    unzip_file(zip_file, folder_name)

print("\nCreated folders:")
for date_label, folder_name in folders.items():
    print(date_label, "->", folder_name)
```

```python
# Cell 4
# Check csv files in each daily folder

for date_label, folder_name in folders.items():
    csv_files = glob.glob(os.path.join(folder_name, "*.csv"))
    
    print("=" * 60)
    print("Date:", date_label)
    print("Folder:", folder_name)
    print("Number of csv files:", len(csv_files))
    print("\nCSV files:")
    
    for file in csv_files:
        print(os.path.basename(file))
```

```python
# Cell 5
# Load all csv files for each day and combine them

def load_one_day(folder_name, date_label):
    csv_files = glob.glob(os.path.join(folder_name, "*.csv"))
    df_list = []
    
    for file in csv_files:
        df = pd.read_csv(file)
        df["source_file"] = os.path.basename(file)
        df["date"] = date_label
        df_list.append(df)
    
    day_df = pd.concat(df_list, ignore_index=True)
    return day_df

day_data = {}

for date_label, folder_name in folders.items():
    day_data[date_label] = load_one_day(folder_name, date_label)

for date_label, df in day_data.items():
    print(date_label, "shape:", df.shape)
```

```python
# Cell 6
# Parse Instrument into underlying, expiry, strike, option_type
# Example: BTC-1MAY26-60000-C

def split_instrument(df):
    df = df.copy()
    
    parts = df["Instrument"].str.split("-", expand=True)
    
    df["underlying"] = parts[0]
    df["expiry"] = parts[1]
    df["strike"] = pd.to_numeric(parts[2], errors="coerce")
    df["option_type"] = parts[3]
    
    return df

for date_label in day_data:
    day_data[date_label] = split_instrument(day_data[date_label])

for date_label, df in day_data.items():
    print("=" * 60)
    print("Date:", date_label)
    print(df[["Instrument", "underlying", "expiry", "strike", "option_type"]].head())
    print("\nMissing strike:", df["strike"].isna().sum())
    print("Option type counts:")
    print(df["option_type"].value_counts(dropna=False))
```

```python
# Cell 7
# Add expiry_date and days_till_maturity

def add_time_to_maturity(df):
    df = df.copy()
    
    df["date"] = pd.to_datetime(df["date"])
    df["expiry_date"] = pd.to_datetime(df["expiry"], format="%d%b%y")
    df["days_till_maturity"] = (df["expiry_date"] - df["date"]).dt.days
    
    return df

for date_label in day_data:
    day_data[date_label] = add_time_to_maturity(day_data[date_label])

for date_label, df in day_data.items():
    print("=" * 60)
    print("Date:", date_label)
    print(df[["Instrument", "date", "expiry", "expiry_date", "days_till_maturity"]].head(10))
    print("\nMinimum days_till_maturity:", df["days_till_maturity"].min())
    print("Maximum days_till_maturity:", df["days_till_maturity"].max())
```

```python
# Cell 8
# Add BTC underlying price for each date

def add_underlying_price(df):
    df = df.copy()
    
    df["date_str"] = df["date"].dt.strftime("%Y-%m-%d")
    df["underlying_price"] = df["date_str"].map(underlying_price_map)
    
    return df

for date_label in day_data:
    day_data[date_label] = add_underlying_price(day_data[date_label])

for date_label, df in day_data.items():
    print("=" * 60)
    print("Date:", date_label)
    print("Missing underlying_price:", df["underlying_price"].isna().sum())
    print(df[["Instrument", "date", "date_str", "strike", "underlying_price"]].head())
    print("Unique underlying_price:", df["underlying_price"].unique())
```

```python
# Cell 9
# Clean price columns
# Replace "-" with NaN and convert to numeric

price_cols = ["Last", "Bid", "Ask", "Mark"]

def clean_price_columns(df):
    df = df.copy()
    
    for col in price_cols:
        df[col] = df[col].replace("-", np.nan)
        df[col] = pd.to_numeric(df[col], errors="coerce")
    
    return df

for date_label in day_data:
    day_data[date_label] = clean_price_columns(day_data[date_label])

for date_label, df in day_data.items():
    print("=" * 60)
    print("Date:", date_label)
    print("Data types:")
    print(df[price_cols].dtypes)
    
    print("\nMissing values:")
    print(df[price_cols].isna().sum())
```

```python
# Cell 10
# Convert option prices from BTC unit to USD unit
# price_USD = price_BTC * underlying_price

price_cols = ["Last", "Bid", "Ask", "Mark"]

def add_usd_price_columns(df):
    df = df.copy()
    
    for col in price_cols:
        usd_col = col + "_USD"
        df[usd_col] = df[col] * df["underlying_price"]
    
    return df

for date_label in day_data:
    day_data[date_label] = add_usd_price_columns(day_data[date_label])

for date_label, df in day_data.items():
    print("=" * 60)
    print("Date:", date_label)
    print(df[[
        "Instrument",
        "strike",
        "underlying_price",
        "Last", "Last_USD",
        "Bid", "Bid_USD",
        "Ask", "Ask_USD",
        "Mark", "Mark_USD"
    ]].head(10))
```

```python
# Cell 11
# Select call options only
# No Madan equation (1), no put-call parity check

call_data = {}

for date_label, df in day_data.items():
    call_df = df[df["option_type"] == "C"].copy()
    
    keep_cols = [
        "Instrument", "date", "date_str", "source_file",
        "underlying", "expiry", "strike", "option_type",
        "expiry_date", "days_till_maturity", "underlying_price",
        "Last", "Bid", "Ask", "Mark",
        "Last_USD", "Bid_USD", "Ask_USD", "Mark_USD"
    ]
    
    call_df = call_df[keep_cols].copy()
    call_data[date_label] = call_df
    
    print("=" * 60)
    print("Date:", date_label)
    print("Call rows:", call_data[date_label].shape)
    print(call_data[date_label][[
        "Instrument", "expiry", "strike", "option_type",
        "underlying_price", "Last", "Last_USD",
        "Bid_USD", "Ask_USD"
    ]].head())
```

```python
# Cell 12
# Drop rows with missing Last
# Last is the target price

last_clean_call_data = {}
removed_missing_last_call_data = {}

for date_label, df in call_data.items():
    before_df = df.copy()
    
    removed_df = before_df[before_df["Last"].isna()].copy()
    clean_df = before_df.dropna(subset=["Last"]).copy()
    
    last_clean_call_data[date_label] = clean_df
    removed_missing_last_call_data[date_label] = removed_df
    
    print("=" * 60)
    print("Date:", date_label)
    print("Before dropping missing Last:", before_df.shape)
    print("After dropping missing Last:", clean_df.shape)
    print("Removed missing Last:", removed_df.shape)
    print("Missing Last after cleaning:", clean_df["Last"].isna().sum())
```

```python
# Cell 13
# Keep OTM call options only
# For call options, OTM means K > S0
# So we keep rows where strike > underlying_price

otm_call_data = {}
removed_not_otm_call_data = {}

for date_label, df in last_clean_call_data.items():
    before_df = df.copy()
    
    clean_df = before_df[before_df["strike"] > before_df["underlying_price"]].copy()
    removed_df = before_df[before_df["strike"] <= before_df["underlying_price"]].copy()
    
    otm_call_data[date_label] = clean_df
    removed_not_otm_call_data[date_label] = removed_df
    
    print("=" * 60)
    print("Date:", date_label)
    print("Before OTM filter:", before_df.shape)
    print("After OTM call filter:", clean_df.shape)
    print("Removed not OTM call:", removed_df.shape)
    print("OTM check, min strike kept:", clean_df["strike"].min() if len(clean_df) > 0 else None)
    print("Underlying price:", before_df["underlying_price"].iloc[0] if len(before_df) > 0 else None)
```

```python
# Cell 14
# Expiry-wise alignment
# For each expiry, keep common strikes across dates
# This makes option curves comparable across dates

all_expiries = sorted(
    set().union(*[set(df["expiry"].unique()) for df in otm_call_data.values()])
)

aligned_parts = {date_label: [] for date_label in otm_call_data.keys()}
alignment_summary = []

for expiry in all_expiries:
    dates_with_expiry = []
    
    for date_label, df in otm_call_data.items():
        if expiry in set(df["expiry"].unique()):
            dates_with_expiry.append(date_label)
    
    common_strikes = None
    
    for date_label in dates_with_expiry:
        one_expiry_df = otm_call_data[date_label][
            otm_call_data[date_label]["expiry"] == expiry
        ]
        
        strikes = set(one_expiry_df["strike"].unique())
        
        if common_strikes is None:
            common_strikes = strikes
        else:
            common_strikes = common_strikes.intersection(strikes)
    
    common_strikes = sorted(common_strikes)
    
    for date_label in dates_with_expiry:
        one_part = otm_call_data[date_label][
            (otm_call_data[date_label]["expiry"] == expiry) &
            (otm_call_data[date_label]["strike"].isin(common_strikes))
        ].copy()
        
        aligned_parts[date_label].append(one_part)
    
    alignment_summary.append({
        "expiry": expiry,
        "number_of_dates": len(dates_with_expiry),
        "dates": ", ".join(dates_with_expiry),
        "common_strikes": len(common_strikes)
    })

aligned_call_data = {}

for date_label in otm_call_data.keys():
    if len(aligned_parts[date_label]) > 0:
        aligned_call_data[date_label] = pd.concat(aligned_parts[date_label], ignore_index=True)
        aligned_call_data[date_label] = aligned_call_data[date_label].sort_values(
            ["expiry_date", "strike"]
        ).reset_index(drop=True)
    else:
        aligned_call_data[date_label] = otm_call_data[date_label].iloc[0:0].copy()

alignment_summary_call_df = pd.DataFrame(alignment_summary)

print("Call no-Madan alignment summary:")
print(alignment_summary_call_df)

print("\nAligned call no-Madan data shapes:")
for date_label, df in aligned_call_data.items():
    print(date_label, "shape:", df.shape)
```

```python
# Cell 15
# Define strict convex cleaning function
# tol = 0
# Call option price should be convex with respect to strike
# This means slopes should be non-decreasing

def find_largest_convex_subset(group, price_col="Last", tol=0):
    group = group.sort_values("strike").copy().reset_index(drop=True)
    
    n = len(group)
    
    if n < 3:
        return group.copy(), group.iloc[0:0].copy()
    
    x = group["strike"].to_numpy(dtype=float)
    y = group[price_col].to_numpy(dtype=float)
    
    dp = np.full((n, n), 2, dtype=int)
    prev_idx = np.full((n, n), -1, dtype=int)
    
    best_len = 2
    best_pair = (0, 1)
    
    for k in range(n):
        for j in range(k):
            slope_jk = (y[k] - y[j]) / (x[k] - x[j])
            
            for i in range(j):
                slope_ij = (y[j] - y[i]) / (x[j] - x[i])
                
                # Strict convex condition:
                # slope should be non-decreasing
                if slope_ij <= slope_jk + tol:
                    if dp[i, j] + 1 > dp[j, k]:
                        dp[j, k] = dp[i, j] + 1
                        prev_idx[j, k] = i
            
            if dp[j, k] > best_len:
                best_len = dp[j, k]
                best_pair = (j, k)
    
    j, k = best_pair
    keep_indices = [k, j]
    
    while prev_idx[j, k] != -1:
        i = prev_idx[j, k]
        keep_indices.append(i)
        k = j
        j = i
    
    keep_indices = sorted(keep_indices)
    
    cleaned_group = group.iloc[keep_indices].copy()
    removed_group = group.drop(index=keep_indices).copy()
    
    return cleaned_group, removed_group


def remove_non_convex_points_one_day(df, price_col="Last", tol=0):
    cleaned_groups = []
    removed_groups = []
    
    for expiry, group in df.groupby("expiry"):
        cleaned_group, removed_group = find_largest_convex_subset(
            group,
            price_col=price_col,
            tol=tol
        )
        
        cleaned_groups.append(cleaned_group)
        
        if len(removed_group) > 0:
            removed_groups.append(removed_group)
    
    cleaned_df = pd.concat(cleaned_groups, ignore_index=True)
    
    if len(removed_groups) > 0:
        removed_df = pd.concat(removed_groups, ignore_index=True)
    else:
        removed_df = pd.DataFrame(columns=df.columns)
    
    return cleaned_df, removed_df

print("Strict convex cleaning functions are ready. tol = 0")
```

```python
# Cell 16
# Run strict convex cleaning
# tol = 0

final_call_data = {}
removed_call_nonconvex_data = {}

for date_label, df in aligned_call_data.items():
    final_df, removed_df = remove_non_convex_points_one_day(
        df,
        price_col="Last",
        tol=0
    )
    
    final_call_data[date_label] = final_df
    removed_call_nonconvex_data[date_label] = removed_df

for date_label in aligned_call_data:
    print("=" * 60)
    print("Date:", date_label)
    print("Before convex cleaning:", aligned_call_data[date_label].shape)
    print("After call no-Madan strict convex cleaning:", final_call_data[date_label].shape)
    print("Removed points:", removed_call_nonconvex_data[date_label].shape)
```

```python
# Cell 17
# Save final strict cleaned call data and removed data
# Version: 10 days, call option, no Madan equation, OTM, tol = 0

print("Saving final cleaned call no-Madan strict data:")

for date_label, df in final_call_data.items():
    file_date = date_label.replace("-", "")
    output_name = f"final_call_no_madan_strict_10days_{file_date}.csv"
    df.to_csv(output_name, index=False)
    print("Saved:", output_name, "| shape:", df.shape)

print("\nSaving removed missing Last call data:")

for date_label, df in removed_missing_last_call_data.items():
    file_date = date_label.replace("-", "")
    output_name = f"removed_call_no_madan_missing_last_10days_{file_date}.csv"
    df.to_csv(output_name, index=False)
    print("Saved:", output_name, "| shape:", df.shape)

print("\nSaving removed not OTM call data:")

for date_label, df in removed_not_otm_call_data.items():
    file_date = date_label.replace("-", "")
    output_name = f"removed_call_no_madan_not_otm_10days_{file_date}.csv"
    df.to_csv(output_name, index=False)
    print("Saved:", output_name, "| shape:", df.shape)

print("\nSaving removed strict non-convex call data:")

for date_label, df in removed_call_nonconvex_data.items():
    file_date = date_label.replace("-", "")
    output_name = f"removed_call_no_madan_strict_nonconvex_10days_{file_date}.csv"
    df.to_csv(output_name, index=False)
    print("Saved:", output_name, "| shape:", df.shape)
```
